```
col = ['baseline value', 'accelerations', 'fetal_movement',
       'uterine_contractions', 'light_decelerations', 'severe_decelerations',
       'prolongued_decelerations', 'abnormal_short_term_variability',
       'mean_value_of_short_term_variability',
       'percentage_of_time_with_abnormal_long_term_variability',
       'mean_value_of_long_term_variability', 'histogram_width',
       'histogram_min', 'histogram_max', 'histogram_number_of_peaks',
       'histogram_number_of_zeroes', 'histogram_mode', 'histogram_mean',
       'histogram_median', 'histogram_variance', 'histogram_tendency']
```

```
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaled=sc.fit_transform(data.drop(['fetal_health'],axis=1))
dt=pd.DataFrame(scaled,columns=col)
dt.head()
```

```
y=data['fetal_health']
x=dt
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```
from sklearn.metrics import classification_report
```

```
from sklearn.linear_model import LogisticRegression
LG=LogisticRegression()
LG.fit(x_train,y_train)
pred_LG=LG.predict(x_test)
acc_LG=LG.score(x_test,y_test)
print(acc_LG)
```

```
print("logistic regression classification report:",classification_report(y_test,pred_LG))
```

```
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier()
DT.fit(x_train,y_train)
pred_DT=DT.predict(x_test)
acc_DT=DT.score(x_test,y_test)
print(acc_DT)
```

```
print("Decision Tree Classifier report:",classification_report(y_test,pred_DT))
```

```
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(n_estimators=100)
RF.fit(x_train,y_train)
pred_RF=RF.predict(x_test)
acc_RF=RF.score(x_test,y_test)
print(acc_RF)
```

```
print("Random Forest Classifier report:",classification_report(y_test,pred_RF))
```

```
from sklearn import svm
SV=svm.SVC(C=2)
SV.fit(x_train,y_train)
pred_SV=SV.predict(x_test)
acc_SV=SV.score(x_test,y_test)
print(acc_SV)
```

```
print("Support Vector Machine Classifier report:",classification_report(y_test,pred_SV))
```

```
from sklearn.ensemble import GradientBoostingClassifier
GBC=GradientBoostingClassifier()
GBC.fit(x_train,y_train)
pred_GBC=GBC.predict(x_test)
acc_GBC=GBC.score(x_test,y_test)
print(acc_GBC)
```

```
print("Gradient Boosting Classifier report:",classification_report(y_test,pred_GBC))
```

```
from xgboost import XGBClassifier
XGC=XGBClassifier()
XGC.fit(x_train,y_train)
pred_XGC=XGC.predict(x_test)
acc_XGC=XGC.score(x_test,y_test)
print(acc_XGC)
```

```
print("XGBoost Classifier report:",classification_report(y_test,pred_XGC))
```

RESULT

```
best_model=pd.DataFrame({"Model":["Logistic Regression","Decision Tree Classifier","Random Forest Classifier","SVM Classifier","Gradient Boosting Classifier","XGBClassifier"],"Accuracy(%)":[acc_LG*100,acc_DT*100,acc_RF*100,acc_SV*100,acc_GBC*100,acc_XGC*100]})
best_model.sort_values(by="Accuracy(%)",ascending=False)
```
